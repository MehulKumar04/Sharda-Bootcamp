## Marital Status and Loans

### Is there a significant difference in the number of personal loans taken by married individuals compared to single individuals?

```python
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(df['Marital_status'], df['Personal_loan'])

# Perform chi-squared test for independence
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print test results
print("Chi-Squared Value:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:")
print(expected)

# Interpret the results
alpha = 0.05  # significance level
if p < alpha:
    print("There is a significant difference between marital status and personal loan uptake.")
else:
    print("There is no significant difference between marital status and personal loan uptake.")
```

## Educational Qualification Impact

### How does educational qualification impact the likelihood of taking a house loan?

```python
contingency_table = pd.crosstab(df['Education'], df['House_loan'])

# Perform chi-squared test for independence
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print test results
print("Chi-Squared Value:", chi2)
print("P-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:")
print(expected)

# Interpret the results
alpha = 0.05  # significance level
if p < alpha:
    print("There is a significant relationship between education and likelihood of taking a house loan.")
else:
    print("There is no significant relationship between education and likelihood of taking a house loan.")
```

## Gender and Salary

### Is there a significant difference in salaries between males and females? Provide statistical evidence.

```python
from scipy import stats
salaries_male = df[df['Gender'] == 'Male']['Salary']
salaries_female = df[df['Gender'] == 'Female']['Salary']

# Perform t-test for independent samples
t_stat, p_val = stats.ttest_ind(salaries_male, salaries_female)

# Print test results
print("T-statistic:", t_stat)
print("P-value:", p_val)

# Interpret the results
alpha = 0.05
if p_val < alpha:
    print("There is a significant difference in salaries between males and females.")
else:
    print("There is no significant difference in salaries between males and females.")
```
